```python
import pandas as pd
import pandera as pa

valores_ausentes = ['**','###!','####',"****","*****",'NULL']
df = pd.read_csv('ocorrencia_2010_2020.csv', sep=';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes) # sep é p dizer como os códigos estão separados
df.head(10)

df.dropna(inplace=True) #Tive que colocar isso pois o #nullable=True em 'ocorrencia_hora' não tava indo
esquema = pa.DataFrameSchema(
    columns = {
        'codigo': pa.Column(pa.Int, required=False),
        'codigo_ocorrencia': pa.Column(pa.Int),
        'codigo_ocorrencia2': pa.Column(pa.Int),
        'ocorrencia_classificacao': pa.Column(pa.String),
        'ocorrencia_cidade': pa.Column(pa.String),
        'ocorrencia_uf': pa.Column(pa.String, nullable=True), #pa.Check.str_len(2,2)
        'ocorrencia_aerodromo': pa.Column(pa.String, nullable=True),
        'ocorrencia_dia': pa.Column(pa.DateTime, nullable=True), # o pa.Check.str_matches tem que vim, sempre, como segundo argumento.
        'ocorrencia_hora': pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True), #validar as horas em 24h
        'total_recomendacoes': pa.Column(pa.Int, nullable=True)
    }
)
```
